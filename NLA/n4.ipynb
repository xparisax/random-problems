Finding eigenvalue for a real tridiagonal symmetric matrix

```python
import numpy as np
from numpy import linalg as LA
```

```python
numbers = input("Pls input 3 numbers(n,diag,main diag): ")
numbers = numbers.split()
n=int(numbers[0]); diag=int(numbers[1]); m_diag=int(numbers[2])
A = np.diag([diag]*(n-1),1) + np.eye(n)*m_diag + np.diag([diag]*(n-1),-1)
```

```python
print(A)
```

```python
eigenvalues = LA.eigvals(A)
print(eigenvalues)
```
